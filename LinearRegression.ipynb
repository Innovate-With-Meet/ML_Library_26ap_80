```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
reg = LinearRegression()
reg2 = LinearRegression()
```

```python
X = [[1], [2], [3], [4], [5], [6]]
Y = [2.5, 2, 2.4, 4.5, 3, 5]  
reg.fit(X,Y)
reg.predict([[5.5]])
```

```python
plt.scatter([x[0] for x in X], Y, color='blue', label='Data Points')
plt.scatter([5.5], reg.predict([[5.5]]), color='red', label="Predicted Point")
plt.plot([x[0] for x in X], reg.predict(X), color='green', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
